```python
import numpy as np
import matplotlib.pyplot as plt
import os

from matplotlib.image import imread # Reading image
```

```python
def plotar_imagem(images, cmap = None, title = None):
  plt.rcParams['figure.figsize'] = [12,12]
  plt.rcParams.update({'font.size': 18})
  fig = plt.figure()
  rows = 1
  cols = 3
  axes=[]

  for i in range(len(images)):
    axes.append(fig.add_subplot(rows, cols, i+1))
    axes[i].set_axis_off()
    plt.imshow(images[i], cmap = cmap)

  fig.tight_layout()
  fig.suptitle(title)
  plt.show  
```

```python
def comprehension(vetor, func):
  lista = list()

  for i in range(len(vetor)):
    if func == np.mean:
      lista.append(func(vetor[i], -1))
    else:
      lista.append(func(vetor[i]))

  return lista
```

```python
imagens = list()

imagens.append(imread('warcraft.jpg'))
imagens.append(imread('dragon_ball.jpg'))
imagens.append(imread('groot.jpg'))
```

```python
imagens_cinzas = np.array(comprehension(imagens, np.mean))
t_image = np.array(comprehension(imagens_cinzas, np.fft.fft2))
t_image_p = np.array(comprehension(comprehension(t_image + 1, np.abs), np.log))
```

```python
plotar_imagem(imagens)
plotar_imagem(imagens_cinzas, cmap = 'gray')
plotar_imagem(t_image_p, cmap='gray')
```
