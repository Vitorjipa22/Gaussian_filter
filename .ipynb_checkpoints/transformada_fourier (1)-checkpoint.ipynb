```python
def plotar_imagem(images, title = None):

  plt.rcParams['figure.figsize'] = [26,26]
  plt.rcParams.update({'font.size': 18})

  fig = plt.figure()
  axes=[]
  n = len(images) / 3

  y = .60
  j = 0
  cmap = None

  # plt.figtext(0.125, y, 'Original image', fontsize = 22, fontname = 'monospace', color = '#283655')
  # plt.figtext(0.4, y, 'Image in gray scale', fontsize = 22, fontname = 'monospace', color = '#283655')
  # plt.figtext(0.675, y, 'Fourier Transform of image', fontsize = 22, fontname = 'monospace', color = '#283655')
 
  for i in range(len(images)):
    
    axes.append(fig.add_subplot(1, n, j+1))
    axes[j].set_axis_off()

    j = j + 1

    plt.imshow(images[i], cmap = cmap)
    cmap = 'gray'
    
    if j == n:
      plt.show
      j = 0
      axes = []
      fig = plt.figure()
```

```python
def comprehension(vetor, func):

  array = list()

  for i in range(len(vetor)):
    
    if func == np.mean or func == np.reshape:
      array.append(func(vetor[i], -1))

    else:
      array.append(func(vetor[i]))

  return np.array(array)
```
